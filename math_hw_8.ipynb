# Level 0

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import expon
from scipy.stats import t
import numpy as np
```

```python
plt.style.use('seaborn-darkgrid')
```

## Ex 1

```python
for scale in [0.6, 0.8, 1, 1.5, 2]:
    _loc = np.random.randint(-2,2)
    
    x = np.linspace(norm.ppf(0.01, loc=_loc, scale=scale), norm.ppf(0.99, loc=_loc, scale=scale), 100)
    plt.plot(x, norm.pdf(x, loc=_loc, scale=scale), label=r'$\mu =${}, $ \sigma^2 =${}'.format(_loc, round(scale**2, 2)))

plt.xlim(-5,5)
plt.legend(frameon=True, framealpha=1, facecolor='white', fontsize=12)

plt.show()
```

```python
x = np.linspace(expon.ppf(0.01), expon.ppf(0.99), 100)
for scale in [0.5, 1, 1.5, 2, 4]:
    
    plt.plot(x, expon.pdf(x, loc=0, scale=scale), label=r'$\lambda =${}'.format(scale))

plt.legend(frameon=True, framealpha=1, facecolor='white', fontsize=12)

plt.show()
```

```python
for df in [1, 1.5, 3, 4, 10]:
    
    x = np.linspace(t.ppf(0.01, df), t.ppf(0.99, df), 100)
    plt.plot(x, t.pdf(x, df=df), label=r'$\nu =${}'.format(df,2))

plt.xlim(-5,5)
plt.legend(frameon=True, framealpha=1, facecolor='white', fontsize=12)

plt.show()
```

```python
def xi_pdf(x):
    a, b = (0, 5)
    if x < a or x > b:
        return 0
    else:
        return 1/(b-a)
    
C = xi_pdf(np.random.randint(0,6))
print(C)
```
